```python
import pandas as pd

# Tworzenie ramki danych z danymi
data = {
    'Respondent': [1, 1, 1, 2, 2, 2, 2, 2, 3, 3, 4, 4, 4, 4],
    'Safety_rate': [2, 2, 1, 4, 3, 2, 1, 1, 2, 4, 3, 2, 1, 1],
    'szerokość_drogi': [5, 5, 1, 1, 5, 5, 1, 1, 1, 5, 1, 5, 5, 1],
    'kolor': ['yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no', 'no', 'no'],
    'bariera_miedzy_drogą': ['yes', 'yes', 'yes', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'yes'],
    'ruch': ['duży', 'duży', 'mały', 'duży', 'mały', 'duży', 'mały', 'mały', 'mały', 'mały', 'duży', 'mały', 'mały', 'mały'],
    'limit_predkości': [50, 50, 30, 30, 30, 30, 50, 30, 50, 50, 50, 30, 30, 50]
}

df = pd.DataFrame(data)

# Obliczenie wskaźnika bezpieczeństwa drogi dla każdego respondent
def calculate_safety_index(row):
    N = row['Safety_rate'].count()  # Liczba ocenionych sytuacji przez danego respondent
    k = 4  # Maksymalna możliwa ocena
    total_score = row['Safety_rate'].sum()  # Suma ocen bezpieczeństwa
    return N * ((k - 1) / total_score) * 100

# Grupowanie danych według respondentów i obliczenie wskaźnika dla każdego z nich
safety_index = df.groupby('Respondent').apply(calculate_safety_index).reset_index(name='Safety_index')

print(safety_index)
```

CHECKING NUMBER OF SITUATION

```python
filtered_df = df[df['profile'].apply(lambda x: x.get('perspective') == 'C')]
```

```python
from collections import Counter
```

```python
def count_scene_ids(rating_list):
    scene_id_counts = Counter()
    for item in rating_list:
        scene_id_counts[item['scene_id']] += 1
    return scene_id_counts

# Zliczamy wystąpienia scene_id w całej kolumnie ratings
total_scene_id_counts = filtered_df['ratings'].apply(count_scene_ids).sum()
```

```python
scene_id_counts_series = pd.Series(total_scene_id_counts)
scene_id_counts_series_sorted = scene_id_counts_series.sort_values(ascending=False)
scene_id_counts_df = scene_id_counts_series_sorted.to_frame(name='count')
print("Count scene_id in ratings columns:")
scene_id_counts_df
```
